```python

import os
import cv2
import numpy as np
import tensorflow
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout


```

```python
animal_list_file = "/content/drive/MyDrive/name_of_the_ animals.txt" # File containing a list of animal names
image_dir = '/content/drive/MyDrive/animals/animals'  # Path to the directory containing animal image folders
```

```python

# Image Resizing
image_size = (224, 224)  # Specify the desired image size
```

```python
def resize_images(image_dir):
    resized_images = []
    for filename in os.listdir(image_dir):
        img = cv2.imread(os.path.join(image_dir, filename))
        img = cv2.resize(img, image_size)
        resized_images.append(img)
    return resized_images
```

```python

with open(animal_list_file, 'r') as f:
    animal_classes = [line.strip() for line in f]
```

```python
X = []
y = []
```

```python
for i, animal_class in enumerate(animal_classes):
    class_dir = os.path.join(image_dir, animal_class)
    images = resize_images(class_dir)
    labels = [i] * len(images)
    X.extend(images)
    y.extend(labels)
```

```python
#  Data Normalization
X = np.array(X, dtype=np.float32)
X /= 255.0
```

```python

# 3. Data Augmentation
datagen = ImageDataGenerator(rotation_range=15, width_shift_range=0.1,
                             height_shift_range=0.1, horizontal_flip=True)
```

```python
# 4. Label Encoding
num_classes = len(animal_classes)
y = to_categorical(y, num_classes)
```

```python

# 5. Build the CNN model
model = Sequential()
model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(image_size[0], image_size[1], 3)))
model.add(MaxPooling2D((2, 2)))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))
model.add(Conv2D(128, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(num_classes, activation='softmax'))
```

```python

# 6. Compile and train the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.fit(datagen.flow(X, y, batch_size=32), epochs=40)

```

```python
# 7. Evaluate the model
loss, accuracy = model.evaluate(X, y)
print("Test Loss:", loss)
print("Test Accuracy:", accuracy)
```

```python
import matplotlib.pyplot as plt

test_image_dir ='/content/drive/MyDrive/test_animals'   # Path to the directory containing test images
```

```python
# Resize and preprocess the test images
test_images = resize_images(test_image_dir)
test_images = np.array(test_images, dtype=np.float32)
test_images /= 255.0
```

```python
# Make predictions on the test images
predictions = model.predict(test_images)
predicted_classes = np.argmax(predictions, axis=1)
```

```python
# Map predicted classes to animal names
predicted_labels = [animal_classes[class_index] for class_index in predicted_classes]

# Print predictions alongside images
fig, axs = plt.subplots(1, len(test_images), figsize=(12, 4))

for i, image in enumerate(test_images):
    axs[i].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    axs[i].axis('off')
    axs[i].set_title(predicted_labels[i])

plt.tight_layout()
plt.show()
```
